### Importing Dependencies


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Data collection and preprocessing

```python
loan= pd.read_csv('/kaggle/input/loan-predication/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
# printing the first five rows of dataset
loan.head()
```

```python
#printing the number of rows and columns
loan.shape
```

```python
#Getting some stats majors from the data
loan.describe()
```

```python
# counting the numbers of missing values
loan.isnull().sum()
```

```python
# dropping the missing values
loan =loan.dropna()
```

```python
# using label encoding 
loan['Loan_Status']=loan['Loan_Status'].replace({'N':0,'Y':1})
```

```python
loan.head()
```

```python
# replacing the 3+ to 4 in dependents columns and changing it into int type
loan['Dependents']=loan['Dependents'].replace({'3+':4}).astype(int)
```

```python
# data visualization
# education and loan status
sns.countplot(x ='Education',hue = 'Loan_Status',data = loan)
```

### observation : the person who is graduate are more capable to get loan

```python
# data visualization
# marital status  and loan status
sns.countplot(x ='Married',hue = 'Loan_Status',data = loan)
```

### observation : the married person have more need then unmarried that's why they have high loan percentage

```python
# data visualization
# Self_Employed and loan status
sns.countplot(x ='Self_Employed',hue = 'Loan_Status',data = loan)
plt.show()
```

### observation : the 	Self_Employed person have unstable income as compare to employed 

```python
# data visualization
# Credit_History and loan status
sns.countplot(x ='Credit_History',hue = 'Loan_Status',data = loan)
plt.show()
```

### observation: 1 = good credit history
###              0 = bad creditistory 
### The person who have good credit history are capable to get loan


```python
# we need to convert all categorical value to numerical value
loan['Married']=loan['Married'].replace({'Yes':1,'No':0}).astype(int)
loan['Self_Employed']=loan['Self_Employed'].replace({'Yes':1,'No':0}).astype(int)
loan['Gender']=loan['Gender'].replace({'Male':1,'Female':0}).astype(int)
loan.replace({'Property_Area': {'Rural': 0, 'Semiurban': 1, 'Urban': 2},
              'Education': {'Graduate': 1, 'Not Graduate': 0}}, inplace=True)
```

```python
loan.head()
```

```python
# seprating the data and label
X = loan.drop(columns= ['Loan_ID','Loan_Status'],axis=1 )
y = loan['Loan_Status']
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.14, random_state=42)
```

```python
# Training the model
model = svm.SVC(kernel='linear')
classifier = model.fit(X_train,y_train)
```

```python
# Model Evaluation
#accuracy score with training data
X_pred =model.predict(X_train)
accuracy = accuracy_score(X_pred,y_train)
print('accuracy on training data',accuracy)
```

```python
# accuracy score with test data
Y_pred = model.predict(X_test)


```

```python
accuracyy = accuracy_score(Y_pred,y_test)
```

```python
print('accuracy on test data',accuracyy)
```
